```python
x = np.linspace(0, 10, num=11, endpoint=True)
x_out = np.arange(-10,20,1)
y = np.cos(-x**2/9.0)
plot1 = plt.plot(x,y,'o',label='original values')
plt.grid()
```

```python
# аппроксимация полиномом
z1 = np.polyfit(x, y, deg=3)
p1= np.poly1d(z1)
```

```python
#  аппроксимация сплайном
from scipy.interpolate import UnivariateSpline
# w = np.isnan(y)
# y[w] = 0.
spl = UnivariateSpline(x, y) #, w=~w)
x_spline = np.linspace(x[0], x[-1], 1000)
x_spline_out = np.linspace(x[0]-10, x[-1]+10, 1000)
```

```python
# интерполяция
from scipy.interpolate import interp1d
f_linear = interp1d(x, y, kind='linear')
f_spline = interp1d(x, y, kind='cubic')

# печать результатов
yvals = p1(x)
yvals_out = p1(x_out)
plot1 = plt.plot(x,y,'o',label='original values')

# plot2 = plt.plot(x,yvals,'r',label='polyfit values')

#plot2 = plt.plot(x_out,yvals_out,'r',label='polyfit values')
plot3 = plt.plot(x_spline,spl(x_spline),'g',label='spline values')

plot4 = plt.plot(x_spline,f_linear(x_spline),'k',label='linear interp')
plot5 = plt.plot(x_spline,f_spline(x_spline),'b',label='cub.spline interp')


plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc=4)
plt.title('polyfitting')
plt.xlim([x[0], x[-1]])
plt.grid()
plt.show()
```
