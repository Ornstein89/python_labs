# Считывание изображения

```python
# вызов справки - Shitf+TAB
# автодополнение и список вариантов ввода - TAB

img = cv2.imread("test_faces2.jpg") # считать массив пикселей из файла
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # перевести цветовое представление из BGR в RGB
matplotlib.pyplot.imshow(img) # отображает в ячейке изобаржение
```

# Настройка OpenCV для распознавания лиц

```python
# алгоритм - каскады Хаара (Haar cascades)

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade # объект, содержащий алгоритм каскадов Хаара, в который загружены обученные коэффициенты
```

# Собственно, само распознавание

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) # перевести изображение img в серое изображение gray_img
matplotlib.pyplot.imshow(gray_img, cmap="gray") # отображает в ячейке изобаржение
matplotlib.pyplot.grid()
```

```python
faces = face_cascade.detectMultiScale(gray_img, 1.1, 4)
# print(faces)
# faces - массив координат прямоугольников [x, y, ширина, высота], содержащих распознанные лица
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2) # функция cv2.rectangle рисует на матрице img прямоугольник с координатами и размерами, цветом RGB=(255, 0, 0)
matplotlib.pyplot.imshow(img)
```
